```python
import pandas as pd
subject_info_df=pd.read_csv("subject_info.csv")
subject_info_df.head()
subject_info_df.describe()
```

```python
import re
intent_name = "Subject_Info"
if intent_name == "Subject_Info":
    # Simulating fetching the subject
    subject = "MCA-102"

    subject_numeric = int(re.search(r'\d+', subject).group())  # Extract numeric part from subject
    print("Subject Numeric:", subject_numeric)
    description_raw = subject_info_df[subject_info_df['Subject Code'] == subject_numeric]['Description'].values[0]
    
    # Printing the raw description for debugging
    print("Raw Description:", description_raw)
    
    # Replace '\\n' with '\n' to ensure proper splitting
    description_raw = description_raw.replace('\\n', '\n')
    
    # Split and format the description
    description_list = description_raw.split('\n')
    print("Description List:", description_list)
    
    formatted_description = []
    for i, concept in enumerate(description_list, start=1):
        formatted_description.append(f"{i}.) {concept.strip()}")  # Format each concept
    
    # Join the formatted descriptions with <br> for HTML
    formatted_description_str = "<br>".join(formatted_description)
    
    # Print the formatted description
    print("Formatted Description String:", formatted_description_str)
```
